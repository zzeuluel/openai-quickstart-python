```python
from settings import *
from pygame image import load
class Preview:
    def __init__(self):
        self.surface = pygame.Surface((SIDEBAR_WIDTH, HEIGHT - PREVIEW_HEIGHT_FRACTION * HEIGHT))
        self.rect = self.surface.get_rect(bottomright=(WINDOW_WIDTH - PADDING, WINDOW_HEIGHT - PADDING))

        # shapes
        self.next_shapes = next_shapes
        #self.shape_surfaces = {shape: load('.../graphics/unnamed.png') for shape, path in TETROMINOS.keys()}
        self.shape_surfaces = {shape: load(path.join('..',graphics,'unnamed.jpg')) for shape, path in TETROMINOS.keys()}

        def display_pieces(self, shapes): 
            for i, shape in enumerate(shapes):
                shape_surface = self.shape_surfaces[shape]
                self.surface.blit(shape_surface, (0, i * 100))  # Adjust position as needed
                x = self.surface.get_width() // 2 - shape_surface.get_width() // 2
                y = self.surface.get_height() // 2 - shape_surface.get_height() //
                self.surface.get_rect(center=(x, y))
                self.surface.blit(shape_surface, (x, y))
    def run(self):
        self.surface.fill(GRAY)
        self.display_pieces(self.next_shapes)
        self.display_surface.blit(self.surface, self.rect)
        pygame.draw.rect(self.display_surface, BLACK, self.rect, 2, 2)
```
