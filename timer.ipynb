```python

class Timer:
    def __init__(self, duration, repeat=False, func=None):
        self.duration = duration
        self.repeated = repeat
        self.func = func

        self.start_time = 0
        self.active = False

    def activate(self):
        self.active = True
        self.start_time = pygame.time.get_ticks()

    def deactivate(self):
        self.active = False
        self.start_time = 0

    def update(self):
        if not self.active:
            return
        current_time = pygame.time.get_ticks()
        if current_time - self.start_time >= self.duration:
            # call a function if provided
            if self.func:
                self.func()
            # repeat or stop
            if self.repeated:
                self.start_time = current_time
            else:
                self.deactivate()
```
